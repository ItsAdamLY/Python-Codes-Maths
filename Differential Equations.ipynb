### Solving ODE using Python ###

```python
#import libraries
import numpy as np
from sympy import *
from scipy.integrate import odeint
from matplotlib import pyplot as plt
```

```python
#declaring symbols
t, I, c= symbols('t I c')

#displaying the separated differential equation
diff_eq = Eq(Integral(1/(5-50*I)), Integral(t**0, t))
diff_eq
```

```python
#integrating the left side
def solve_left(I):
    z1 = integrate(1/(5-50*I))
    return z1

solve_left(I)
```

```python
#integrating the right side
def solve_right(t):
    z2 = integrate(t**0, t) + c #putting +c in the end because it is an indefinite integral
    return z2

solve_right(t)
```

```python
#explicitly solving for I by using solve() function
z = solve(solve_left(I) - solve_right(t),I)
init_printing(z)
z
```

### Using Python to Plot the Solutions ###

```python
#defining the differential equation model in a function
def equation_model(I, t):
    dIdt = 5 - 50*I
    return dIdt

def solved_equation(c):
    I = 1/10 - c*np.exp(-50*t)
    return I
```

```python
#setting initial values
t = np.linspace(0,0.5)
I0 = 0

#set one variable for Python's differential equation calculation (with initial values)
w = odeint(equation_model, I0, t)

#set another variable for manually calculated differential equation (with C = 1/10)
y = solved_equation(1/10)

#setting axes properties
fig, axis = plt.subplots()
axis.set(title="Graphs of solutions for the ODE", xlabel="t (seconds)", ylabel="I (Amperes)")
axis.grid()

#plotting the graph using matplotlib.pyplot
plt.plot(t,w,linewidth=3, color="#00a8bb", label="Solution using Python")
plt.plot(t,y,linestyle='dashed',linewidth=3,color="#007800", label="Solution using Manual Calculations")

legend = axis.legend(loc="right",shadow=True)

plt.show()
```
